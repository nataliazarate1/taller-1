## Tabla Vivienda de la Encuesta Nacional de Calidad de Vida - ECV - 2024

```python
# Traer archivo del DANE
ruta_csv = "/home/jovyan/datos_dane/DBF-ENCV-Datos_vivienda-2024/Datos de la vivienda.CSV"

datos= pd.read_csv(ruta_csv, sep=';', encoding='latin-1')  # sep=';' para CSV con punto y coma
print(datos.info())
print(datos.head())
```

```python
# Columnas identificadores y geográficas
identificadores = ["DIRECTORIO", "SECUENCIA_ENCUESTA", "SECUENCIA_P", "ORDEN"]
geograficas = ["P1_DEPARTAMENTO", "P1_MUNICIPIO", "REGION"]

# Columnas numéricas para estadísticas
numericas = ["CANT_HOG_COMPLETOS", "CANT_HOGARES_VIVIENDA", "FEX_C"]

# Columnas categóricas (ordenadas o nominales)
categoricas = [c for c in datos.columns if c not in identificadores + geograficas + numericas]

# Convertir tipos correctamente
for c in geograficas + categoricas:
    datos[c] = datos[c].astype("category")
```

## Análisis descriptivo

### Variables numéricas

```python
# Estadísticas descriptivas
estadisticas_num = datos[numericas].describe()
estadisticas_num.to_csv("/home/jovyan/resultados/estadisticas_numericas.csv")
print(estadisticas_num)
```

### Variables categóricas

```python
estadisticas_cat = {}
for c in geograficas + categoricas:
    conteos = datos[c].value_counts(dropna=False)
    porcentajes = datos[c].value_counts(normalize=True, dropna=False) * 100
    df_temp = pd.DataFrame({"conteo": conteos, "porcentaje": porcentajes})
    df_temp.to_csv(f"/home/jovyan/resultados/conteos_{c}.csv")
    estadisticas_cat[c] = df_temp
```

### Análisis servicios públicos

```python
# Variables de servicios públicos
servicios_publicos = {
    'P8520S1': 'Energia_electrica',
    'P8520S1A1': 'Estrato_tarifa',
    'P8520S5': 'Acueducto',
    'P8520S3': 'Alcantarillado',
    'P8520S4': 'Recoleccion_basuras'
}

# Función de análisis para variables categóricas
def analisis_servicios(df, variables_dict):
    for var, nombre in variables_dict.items():
        print(f"\n=== Análisis de {nombre} ===")
        df[var] = df[var].replace({1:'Si', 2:'No'})
        
        # Estadísticas generales
        conteos = df[var].value_counts(dropna=False)
        porcentajes = df[var].value_counts(normalize=True, dropna=False) * 100
        df_estadisticas = pd.DataFrame({"conteo": conteos, "porcentaje": porcentajes})
        print(df_estadisticas)
        
        # Tabla cruzada por departamento
        tabla_depto = pd.crosstab(df['P1_DEPARTAMENTO'], df[var], margins=True, normalize='index')*100
        print(tabla_depto)
        
        # Tabla cruzada por región
        tabla_region = pd.crosstab(df['REGION'], df[var], margins=True, normalize='index')*100
        print(tabla_region)
        
        # Gráfico de barras
        plt.figure(figsize=(6,4))
        df[var].value_counts().plot(kind='bar', color='skyblue')
        plt.title(f"Distribución de {nombre}")
        plt.ylabel("Cantidad de viviendas")
        plt.xlabel("Respuesta")
        plt.tight_layout()
        plt.savefig(f"resultados/grafico_{nombre}.png")
        plt.show
analisis_servicios(df, servicios_publicos)
```
